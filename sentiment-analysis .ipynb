# import the dataset

```python
df_alexa = pd.read_csv("amazon_alexa_data.csv")
df_alexa.head()
```

```python
df_alexa.drop(df_alexa.columns[df_alexa.columns.str.contains('Unnamed',case = False)],axis = 1, inplace = True)

df_alexa.head()
```

```python
from sklearn.preprocessing import OneHotEncoder
x_cat = df_alexa[['variation']]
encoder = OneHotEncoder()
x_cat = encoder.fit_transform(x_cat).toarray()
x_cat.shape
```

```python

df2  = [df_alexa['verified_reviews'], df_alexa['feedback'], x_cat]
df_alexa= pd.concat([df_alexa,x_cat], axis=1)
df_alexa.drop(['variation'],axis=1,inplace=True)
df_alexa.head()
```

# tokenization of reviews

## import nltk for tokenization  

### convert  to lower case 

### remove punctuation

```python
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords 
import string

df= np.asarray(df_alexa['verified_reviews'])
df_token=[]
for i in df:
    i= i.lower()  #convert  to lower case
    i = i.translate(str.maketrans('','',string.punctuation))   #removed punctuation 
    df_token.append(nltk.word_tokenize(i)) 

```

### stopwords removed

### lemmatizing words

```python
from nltk.stem import WordNetLemmatizer

nltk.download('wordnet')
nltk.download('omw-1.4')

stop_words = list(stopwords.words('english'))
df_token_filter = []

lemmatizer = WordNetLemmatizer()

for i in df_token:
    i = [x for x in i if x not in stop_words]
    i = [lemmatizer.lemmatize(x) for x in i]
    i = ' '.join(i)
    df_token_filter.append(i)
print(df_token_filter)
```

```python
df_alexa['verified_reviews'] = df_token_filter
```

```python
df_alexa.head()
```

## transform words to vectors

```python
from sklearn.feature_extraction.text import CountVectorizer
countv = CountVectorizer()

reviews_countv = countv.fit_transform(df_alexa['verified_reviews'])

reviews_countv.shape
```

```python
df_alexa.drop(['verified_reviews'], inplace =True , axis =1)
```

```python
df_alexa.head()
```

```python
df_alexa2 = pd.DataFrame(reviews_countv.toarray())
df_alexa2
```

### determine X and y

```python
X = df_alexa2
y = df_alexa['feedback']
```

### split test, train dataset

```python
from sklearn.model_selection import train_test_split as tts

X_train,X_test, y_train , y_test = tts(X , y , test_size = 0.2)
print(X_train.shape, y_train.shape)
```

# implement Multinomial Na√Øve Bayes Classification

```python
from sklearn.naive_bayes import MultinomialNB
clf = MultinomialNB().fit(X_train, y_train)

predicted = clf.predict(X_test)
predicted
```

```python
from sklearn.metrics import accuracy_score as ac
from sklearn.metrics import confusion_matrix as cm

ac_mnb = ac(y_test, predicted)

print("accuracy score: " ,ac_mnb)
print("confusion matrix:  ")

cm(y_test, predicted)
```

# implement logistic regression

```python
from sklearn.linear_model import LogisticRegression as lr
regr = lr()
regr = regr.fit(X_train,y_train)
```

```python
predicted_lr = regr.predict(X_test)
predicted_lr
```

```python
ac_lr = ac(y_test, predicted_lr)

print("accuracy score: ", ac_lr)
print("confussion matrix: ")

cm(y_test, predicted_lr)
```

# implement KNN classifiers

```python
from sklearn.neighbors import KNeighborsClassifier 
```

#### check how many neighbor produce best accuracy

```python
Ks = 12
mean_acc = np.zeros((Ks-1))
std_acc = np.zeros((Ks-1))

for n in range(1,Ks):
    
    #Train Model and Predict  
    neigh = KNeighborsClassifier(n_neighbors = n).fit(X_train,y_train)
    yhat=neigh.predict(X_test)
    mean_acc[n-1] = ac(y_test, yhat)

    
    std_acc[n-1]=np.std(yhat==y_test)/np.sqrt(yhat.shape[0])

mean_acc
```

```python
# from above test, neighbors = 5 looks optimal,so let's use it
knn = KNeighborsClassifier(n_neighbors = 5).fit(X_train,y_train)
predicted_knn = knn.predict(X_test)

ac_knn = ac(y_test ,predicted_knn)

print("accuracy score: ",ac_knn)
print("confusion matrix:")

cm(y_test, predicted_knn)
```

### compare all classification modal accuracy

```python
print("multinomialNB classifier test accuracy: ",ac_mnb)
print("logistic regression test accuracy: ",ac_lr)
print("K neighbour classifier test accuracy: ",ac_knn)    
```

### compare all confusion matrix

```python
print("\nmultinomialNB classifier confision matrix: ")
print(cm(y_test,predicted))
print("\nlogistic regression confision matrix: ")
print(cm(y_test,predicted_lr))
print("\nK neighbour classifier confision matrix: ")
print(cm(y_test,predicted_knn))
```
